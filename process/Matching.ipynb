```python
path_booking = '/content/drive/MyDrive/HUST/20202/Tích hợp dữ liệu/booking.com/booking.json'
path_hanoi = '/content/drive/MyDrive/HUST/20202/Tích hợp dữ liệu/data/traveloka/ha_noi.csv'
```

```python
import pandas as pd

booking_full = pd.read_json(path_booking)
booking_full.T
```

```python
hanoi_full = pd.read_csv(path_hanoi)
hanoi_full.T
```

```python
cols = ['hotel name', 'rating', 'stars', 'n_reviews']
booking = booking_full[cols]

def convert2num(x):
  if x and (type(x) == str):
    res = ""
    for i in range(len(x)):
      if x[i] == ',':
        res += '.'
      else:
        res += x[i]
    return res
  return x

booking['rating'] = booking['rating'].apply(convert2num).astype(float)
booking['stars'] = booking['stars'].apply(convert2num).astype(float)
booking['n_reviews'] = booking['n_reviews'].apply(convert2num).astype(float)
```
