```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML
```

# Load data

```python
path = {
    'booking': '../bookingv2.csv',
    'agoda': '../agoda5.json',
    'traveloka_hanoi': '../traveloka/ha_noi.json',
    'traveloka_danang': '../traveloka/da_nang.json',
    'traveloka_hcm': '../traveloka/ho_chi_minh.json'
}
```

```python
booking_df = pd.read_csv(path['booking'])
agoda_df = pd.read_json(path['agoda'])

tra_hanoi_df = pd.read_json(path['traveloka_hanoi'])
tra_danang_df = pd.read_json(path['traveloka_danang'])
tra_hcm_df = pd.read_json(path['traveloka_hcm'])
```

```python
# merge 3 df traveloka to 1 traveloka_df
tra_hanoi_df['city'] = 'Hà Nội'
tra_danang_df['city'] = 'Đà Nẵng'
tra_hcm_df['city'] = 'Hồ Chí Minh'
traveloka_df = pd.concat([tra_hcm_df, tra_hanoi_df, tra_danang_df], axis=0)
traveloka_df.reset_index(drop=True, inplace=True)
traveloka_df
```

```python
booking_df.drop(booking_df.columns[0], axis=1, inplace=True)
booking_df
```
